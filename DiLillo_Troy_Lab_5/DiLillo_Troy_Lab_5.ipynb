```python
#Postlab 5
#First part make sure to comment on it 
#These imports allow us to use the tools in the future program
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC

#Allows our program to launch a specified website through google chrome
PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service = PATH)

#The website that will be launched through chrome
driver.get("https://www.techwithtim.net/" )

#specifies the link that we will be clicking to access a different part of the page. Using what is visible to us.
link = driver.find_element(By.LINK_TEXT,"Python Programming")
#Clicking the link
link.click()

#try loop tells us if something is going wrong by sending a signal of our choosing. In this case it will quit.
try:
    
    #We tell our program to slow down so it doesn't trip over its self. We also tell it to wait for a given element
    #We then tell our program to click said element.
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials")))
    element.click()
   
    #Repeat of the previous two comments except we tell our program to look out for an ID rather than a link.
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "sow-button-19310003")))
    element.click()
  
    #these are the actions telling our prgram to move around the pages
    driver.back()
    driver.back()
    driver.back()
    driver.forward()
    driver.forward()
#except as mentioned before will quit the program if it is unable to do what we previously instructed it
except:
    driver.quit()

```

```python
PATH = Service("C:/Program Files (x86)/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://data.gov/")

link = driver.find_element(By.LINK_TEXT,"DATA")
link.click()

search = driver.find_element(By.NAME, "q")
search.clear() #clear any text in the search just in case. Not needed here.
search.send_keys("climate change")
search.send_keys(Keys.RETURN)

csv_file = open("Climate Change", "w", newline = "", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title", "Description"])



#use an if statement to break it when it reaches 21 it won't work

try:
    main = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "module-content")))
    articles = main.find_elements(By.TAG_NAME, "li")
   
    for i,element in enumerate(articles):
        if i+1 <= 20:
            title = element.find_element(By.XPATH, f"//*[@id='content']/div[2]/div/section[1]/div[2]/ul/li[{i+1}]/div/h3/a").text
       
            entry = element.find_element(By.XPATH, f"//*[@id='content']/div[2]/div/section[1]/div[2]/ul/li[{i+1}]/div/div[2]").text
        
            print(title,entry)
            csv_writer.writerow([title, entry])
        else:
            break
        
        


        
    
       

finally:
    driver.quit()      


csv_file.close()
```
